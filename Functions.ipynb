## Functions
The following shows how functions are done in calculating Tax in Zimbabwe


```python
# Daily tax
def daily_tax(x):
    if x <= 2.3:
        z = x* 0    # 0% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))
    elif x > 2.3 and x <= 9.86:
        z = x*0.2 - 0.46  # 20% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))
    elif x > 9.86 and x <= 32.88:
        z = x*0.25 - 0.95   # 25% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))
    elif x > 32.88 and x <= 65.75:
        z = x*0.3 - 2.6  # 30% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))
    elif x > 65.77 and x <= 98.63:
        z = x*0.35 - 5.88  # 35% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))
    else:
        z = x*0.40 - 10.82  # 40% Tax
        y = x - z
        print('Your tax is ${} your net wage is ${}'.format(round(z, 2), round(y, 2)))

daily_tax(float(input("What is your daily wage?:")))
```
